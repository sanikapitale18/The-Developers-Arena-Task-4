```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_sales = {
    "Category": ["Electronics", "Clothing", "Groceries", "Furniture", "Toys"],
    "Sales":    [85000,        54000,       73000,       42000,      31000]
}

df_sales = pd.DataFrame(data_sales)
print(df_sales)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_sales["Category"], df_sales["Sales"])

plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Sales (in â‚¹)")
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```python
data_expenses = {
    "Category": ["Rent", "Groceries", "Transport", "Utilities", "Entertainment", "Other"],
    "Amount":   [15000, 8000, 2500, 3000, 2000, 1500]
}

df_exp = pd.DataFrame(data_expenses)
print(df_exp)
```

```python
import pandas as pd

data_students = {
    "Student": ["Gauri", "Parnashree", "Sanika", "Samiya", "Hridya", "Chetna", "Sharvari", "Rudransh"],
    "Maths": [78, 92, 65, 88, 55, 80, 73, 95],
    "Science": [82, 89, 60, 90, 58, 77, 70, 93],
    "English": [75, 85, 68, 91, 62, 79, 72, 88]
}

df = pd.DataFrame(data_students)

df.to_csv("students.csv", index=False)
```

```python
df = pd.read_csv("students.csv")
print(df)
print(df.columns.tolist())
```

```python
df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
df["Average"] = df["Total"] / 3
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 60 else "Fail")

print(df)
```

```python
subject_means = df[["Maths", "Science", "English"]].mean()

plt.figure(figsize=(6, 4))
plt.bar(subject_means.index, subject_means.values)

plt.title("Average Marks by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Marks")

plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
result_counts = df["Result"].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(result_counts.values,
        labels=result_counts.index,
        autopct="%1.0f%%",
        startangle=90)

plt.title("Pass vs Fail Distribution")
plt.tight_layout()
plt.show()
```
